```
import os
import string
import pandas as pd
import numpy as np
import re
import pprint
import ast
from urllib.request import urlopen

import requests
from google.colab.output import eval_js
from IPython.display import Javascript

from IPython.display import Image
from sklearn.manifold import TSNE

!git clone https://github.com/CAHLR/d3-scatterplot.git

pokemon = pd.read_csv("/content/drive/MyDrive/DATA 144 Final Project/merged_pokemon.csv", index_col = 0)
pokemon["Abilities"] = [ast.literal_eval(x) for x in pokemon["Abilities"]]

all_abilities = set()
for i in pokemon["Abilities"]:
  for ability in i:
    all_abilities.add(ability)
print(len(all_abilities))
for ability in all_abilities:
  pokemon[ability] = [(ability in abilities) for abilities in pokemon["Abilities"]]   

pokemon["Generation"] = pokemon["Generation"].astype(str)
pokemon_plot = pokemon.drop(columns = ["#",
                                       "Name", 
                                       "Abilities", 
                                       "Type 1", 
                                       "Type 2", 
                                       "Generation", 
                                       "Experience type",
                                       "Experience to level 100", 
                                       "Legendary", 
                                       "Against Normal",
                                       "Against Fire",
                                       "Against Water",
                                       "Against Electric",
                                       "Against Grass",
                                       "Against Ice",
                                       "Against Fighting",
                                       "Against Poison",
                                       "Against Ground",
                                       "Against Flying",
                                       "Against Psychic",
                                       "Against Bug",
                                       "Against Rock",
                                       "Against Ghost",
                                       "Against Dragon",
                                       "Against Dark",
                                       "Against Steel",
                                       "Against Fairy",
                                       "Final Evolution",
                                       "Mega Evolution",
                                       "Alolan Form",
                                       "Galarian Form"])
pokemon_plot

pokemon_features = pokemon[["Name",
                            "Mean",
                            "Standard Deviation",
                            "HP",	
                            "Att",	
                            "Def",	
                            "Spa",	
                            "Spd",	
                            "Spe",	
                            "BST", 
                            "Type 1", 
                            "Type 2", 
                            "Generation", 
                            "Experience type", 
                            "Legendary", 
                            "Catch Rate", 
                            "Final Evolution", 
                            "Mega Evolution"]]

from sklearn.manifold import TSNE

data_embed=TSNE(n_components=2, perplexity=50, verbose=2, method='barnes_hut', random_state = 10).fit_transform(pokemon_plot)


df = pd.DataFrame(data_embed ,columns=["x","y"]).merge(pokemon_features, left_index = True, right_index = True)

df.to_csv('d3-scatterplot/mytext.tsv', sep='\t', index=False)

!git clone https://github.com/CAHLR/d3-scatterplot.git
df.to_csv('d3-scatterplot/mytext.tsv', sep='\t', index=False)

def show_port(port, data_file, width=600, height=800):
  display(Javascript("""
  (async ()=>{
    fm = document.createElement('iframe')
    fm.src = await google.colab.kernel.proxyPort(%d) + '/index.html?dataset=%s'
    fm.width = '90%%'
    fm.height = '%d'
    fm.frameBorder = 0
    document.body.append(fm)
  })();
  """ % (port, data_file, height)))

port = 8000
data_file = 'mytext.tsv'
height = 1400

get_ipython().system_raw('cd d3-scatterplot && python3 -m http.server %d &' % port) 
show_port(port, data_file, height)

# legendary is grouped together 
```
