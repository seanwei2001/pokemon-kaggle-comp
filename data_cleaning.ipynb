```
import numpy as np
import pandas as pd
```

```
pokemon_raw= pd.read_csv('All_Pokemon.csv')
print(pokemon_raw.shape)
pokemon_raw.head()
```

```
pokemon_combat_id = pd.read_csv('pokemon_combat_id.csv')[['#', 'Name']]
print(pokemon_combat_id.shape)
pokemon_combat_id.head()
```

```
pokemon = pokemon_raw.merge(pokemon_combat_id, on = 'Name', how = 'left').rename(columns = {'#' : 'Combat_ID'})
pokemon.head()
```

```
missing_pokemon = {}
for i in pokemon_combat_id['#']:
    if i not in list(pokemon['Combat_ID']):
        missing_pokemon[i] = (list(pokemon_combat_id[pokemon_combat_id['#'] == i]['Name'])[0])

print(len(missing_pokemon))
missing_pokemon
```

```
pokemon_combat_id_replaced = pokemon_combat_id.replace(change_this_shit)
```

```
pokemon_inner_join = pokemon_combat_id_replaced.merge(pokemon_raw, how = 'inner', on = 'Name')
pokemon_inner_join
# Inner join causes it to miss 2 because 2 meowth version turns into 1 and 2 Keldeo versions turn into 1
```

```
pokemon = pokemon_combat_id_replaced.merge(pokemon_raw, how = 'left', on = 'Name').drop(columns = ["Number"])
pokemon
#left join fixes this
```

```
pokemon.to_csv("merged_pokemon.csv")
```
